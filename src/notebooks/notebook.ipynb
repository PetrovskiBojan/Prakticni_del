```python
import pandas as pd
import os
```

```python
csv_df = pd.read_csv("../../data/raw/students_scores.csv")

print("CSV DataFrame:")
print(csv_df.head()) 
```

```python
xlsx_df = pd.read_excel("../../data/raw/students.xlsx")

print("Excel DataFrame:")
print(xlsx_df.head())
```

```python
df = pd.concat([csv_df, xlsx_df], axis=1)
df.head()
```

```python
df = df.drop(columns=["Unnamed: 0"])
df.head()
```

```python
df["MIN_subject"] = df[["STEM_subjects", "H_subjects"]].min(axis=1)
print(df["MIN_subject"].head())
```

```python
output_path = "../../data/processed/current_data.csv"
os.makedirs(os.path.dirname(output_path), exist_ok=True)

df.to_csv(output_path, index=False)
```
